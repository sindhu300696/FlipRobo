```python
from bs4 import BeautifulSoup
import requests
```

```python
page1=requests.get("https://www.imdb.com/chart/top/")
page1.content
```

```python
soup=BeautifulSoup(page1.content,'html.parser')
print(soup.prettify())
```

```python
title_list=[]
years_list=[]
rating_list=[]
```

```python
title= soup.find_all('td', class_ = "titleColumn")

for i in title:
    title_list.append(i.get_text().replace('\n',' '))
title_list[0:5]
```

```python
years = soup.find_all('span', class_="secondaryInfo")

for i in years:
    years_list.append(i.get_text().replace('\n',' '))
years_list[0:5]
```

```python
ratings = soup.find_all('td', class_="ratingColumn imdbRating")

for i in ratings:
    rating_list.append(i.get_text().replace('\n',' '))
rating_list[0:5]
```

```python
import pandas as pd
movies=pd.DataFrame({})
movies['Title']=title_list[:100]
movies['Year']=years_list[:100]
movies['Rating']=rating_list[:100]
```

```python
movies['Title']=movies['Title'].str.split("(").str[0]
```

```python
movies
```

```python
movies.to_csv('file1.csv') 
```
