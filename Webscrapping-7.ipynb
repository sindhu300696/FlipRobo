```python
from bs4 import BeautifulSoup
import requests
```

```python
page2=requests.get("https://www.amazon.in/s?k=mobiles+under+20000&ref=nb_sb_noss_2")
page2.content
```

```python
soup=BeautifulSoup(page2.content,'html.parser')
print(soup.prettify())
```

```python
name_list=[]
price_list=[]
rating_list=[]
image_list=[]
```

```python
name= soup.find_all('span', class_ = "a-size-medium a-color-base a-text-normal")

for i in name:
    name_list.append(i.get_text().replace('\n',' '))
name_list[0:5]
```

```python
price= soup.find_all('span', class_ = "a-price-whole")

for i in price:
    price_list.append(i.get_text().replace('\n',' '))
price_list[0:5]
```

```python
rating= soup.find_all('i', class_ = "a-icon a-icon-star-small a-star-small-4 aok-align-bottom")

for i in rating:
    rating_list.append(i.get_text().replace('\n',' '))
rating_list[0:5]
```

```python
import urllib.request
images = soup.find_all('img') 
for item in images: 
    print(item['src']) 
```

```python
for i in images:
    image_list.append(item['src'])
image_list[0:5]
```

```python
import pandas as pd
p=pd.DataFrame({})
p['Name']=name_list[:14]
p['Price']=price_list[:14]
p['Rating']=rating_list[:14]
p['Image URL']=image_list[:14]
```

```python
p['Name']=p['Name'].str.split("(").str[0]
```
