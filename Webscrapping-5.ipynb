```python
from bs4 import BeautifulSoup
import requests
```

```python
page1=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/test")
page1.content
```

```python
soup=BeautifulSoup(page1.content,'html.parser')
print(soup.prettify())
```

```python
team_list=[]
match_list=[]
points_list=[]
rating_list=[]
```

```python
team= soup.find_all('span', class_ = "u-hide-phablet")

for i in team:
    team_list.append(i.get_text())
team_list[0:5]
```

```python
points= soup.find_all('td', class_ = "table-body__cell u-center-text")

for i in points:
    points_list.append(i.get_text())
points_list[0:5]
```

```python
match= soup.find_all('td', class_ = "table-body__cell u-center-text")
for i in match:
    match_list.append(i.get_text())
match_list[0:5]
```

```python
rating= soup.find_all('td', class_ = "table-body__cell u-text-right rating")

for i in rating:
    rating_list.append(i.get_text().replace('\n',' '))
rating_list[0:5]
```

```python
import pandas as pd
cricket=pd.DataFrame({})
cricket['Team']=team_list[:8]
cricket['Points']=points_list[:8]
cricket['Match']=match_list[:8]
cricket['Rating']=rating_list[:8]
```

```python
cricket
```

```python
page2=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi")
page2.content
```

```python
team_bat= soup.find_all('span', class_ = "table-body__logo-text")

for i in team_bat:
    team3_bat_list.append(i.get_text())
team3_bat_list[0:5]
```

```python
rating_bat= soup.find_all('td', class_ = "table-body__cell u-text-right rating")
for i in rating_bat:
    rating_bat_list.append(i.get_text().replace('\n',' '))
rating_bat_list[0:5]
```

```python
import pandas as pd
final=pd.DataFrame({})
final['Name']= name_bat_list[:10]
final['Team']=team3_bat_list[:10]
final['Rating']=rating_bat_list[:10]
```
