```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
name_list=[] 
author_list=[]
genre_list=[]
review_list=[]
```

```python
name= soup.find_all('h4', class_ = "italic")

for i in name:
    name_list.append(i.get_text().replace('\n',' '))
name_list[0:5]
```

```python
author= soup.find_all('p', class_ = "sans bold")

for i in author:
    author_list.append(i.get_text().replace('\n',' '))
author_list[0:5]
```

```python
genre= soup.find_all('p', class_ = "genre-links hidden-phone")

for i in genre:
    genre_list.append(i.get_text().replace('\n',' '))
genre_list[0:5]
```

```python
review= soup.find_all('p', class_ = "excerpt")

for i in review:
    review_list.append(i.get_text().replace('\n',' '))
review_list[0:5]
```

```python
import pandas as pd
book=pd.DataFrame({})
book['Name']=name_list[:5]
book['Author']=author_list[:5]
book['Genre']=genre_list[:5]
book['Review']=review_list[:5]
```
